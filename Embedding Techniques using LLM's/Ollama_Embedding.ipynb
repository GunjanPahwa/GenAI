#### OLLAMA
1. This is opensource, use if you dont have access to openai
2. Ollama is a platform where we can use opensource LLm models like llama, mistral, gemma
3. Ollama supports embedding models, making it possible to build retrieval augmented generation(RAG) applications that combine text prompts with existing documents or other data


```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="llama3") #by default it uses llama2, i downloaded llama3 using cmd
) #download the model by seeing the opensource documentation of ollama
```

```python
embeddings
```

```python
r1=embeddings.embed_documents([
    "Alpha is the first letter of the Greek alphabet",
    "Beta is the second letter of the Freek alphabet",
])
```

```python
r1
```

```python
len(r1[0]) #so llama3 uses 4096 to represent a single word in form of vectors
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet") #this will match r1[1]
```

```python
#Other embedding models
embeddings=OllamaEmbeddings(model="mxbai-embed-large")
text="This is a test document"
query_result=embeddings.embed_query(text)
```

```python
len(query_result) #1024 dimensions
```
