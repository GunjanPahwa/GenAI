```python
#ALL OF THE FOLLOWING CAN BE DONE ONLY WITH OPENAI KEY
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large") #choose the model according to the work you need it for
embeddings
```

```python
text="This is a tutorial on OpenAI embedding"
query_result=embeddings.embed_query(text) #converts text to vectors
query_resukt
```

```python
len(query_result) #3072 dimensions
#this model takes a sentence and converts it into a vector of 3072 dimensions
```

```python
#I can also decide the dimensions i want, say i do not want 3072 dimensions
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimension=1024) #this means convert my sentence into a vector of 1024 dimensions
embeddings_1024
```

```python
text="This is a tutorial on OpenAI Embedding"
query_result=embeddings_1024.embed_query(text)
len(query_result) #1024 dimensions for same text
```

```python
#converting speech.txt into vectors
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python

from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs=text_splitter.split_documents(docs)
final_docs
```

```python
#I took some text, converted it into chunks and now i will convert it to vectors and store it to VectorStoreDB
# Vector Embedding and Vector Store
from langchain_community.vectorstores import Chroma #We are using Chrona Database
```

```python
db=Chroma.from_documents(final_docs,embeddings_1024) #the document to be converted and how to do the embedding
db
```

```python
#how to query from this database
query="It will be all the easier for us to conduct ourselves as belligerents" #took this line from speech.txt
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
